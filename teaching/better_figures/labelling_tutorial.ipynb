#Labelling Tutorial

```python
###Import modules
import matplotlib.pyplot
import numpy

%matplotlib inline
```

```python
###Generate some fake data
###Lines
##numpy.arange(start, end, step) generates a series with end not inclusive
x_line = numpy.arange(0.0, 10.0, 0.5)
y_line1 = x_line**2 + 2
y_line2 = numpy.sqrt(x_line) + 3
y_line3 = numpy.cos(x_line)
y_line4 = numpy.sin(x_line)

###Scatter
##Generate N random points
N = 100
x_scatter = numpy.random.rand(N)
y_scatter = numpy.random.rand(N)
z_scatter = numpy.random.rand(N)
```

```python
###Plot line plots: Different ways of labelling/legend
matplotlib.pyplot.figure(figsize=(10,10)) ##width, height

ax = matplotlib.pyplot.subplot(3,2,1) ##rows, columns, plot
ax.plot(x_line, y_line1, 'k-o', lw=3, label='Line one')
ax.plot(x_line, y_line2, 'k--s', lw=3, label='Line two')
ax.legend()
ax.set_title('Labelling with Legend, with Frame')
#ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax = matplotlib.pyplot.subplot(3,2,2) ##rows, columns, plot
ax.plot(x_line, y_line1, 'k-o', lw=3, label='Line one')
ax.plot(x_line, y_line2, 'k--s', lw=3, label='Line two')
ax.legend(frameon=False)
ax.set_title('Labelling with Legend, no Frame')
#ax.set_xlabel('x axis')
#ax.set_ylabel('y axis')

ax = matplotlib.pyplot.subplot(3,2,3) ##rows, columns, plot
ax.plot(x_line, y_line3, 'b-', lw=3)
ax.plot(x_line, y_line4, 'g-', lw=3)
ax.text(2.7, 0.75, 'Line three', color='b', fontsize=14)
ax.text(2.7, 0.5, 'Line four', color='g', fontsize=14)
ax.set_title('Labelling with Text')
#ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax = matplotlib.pyplot.subplot(3,2,4) ##rows, columns, plot
ax.axis([0,14,-1.1,1.1]) ## [xmin, xmax, ymin, ymax]
ax.plot(x_line, y_line3, 'b-', lw=3)
ax.plot(x_line, y_line4, 'g-', lw=3)
ax.text(x_line[-1]+0.25, y_line3[-1], 'Line three', color='b', fontsize=14)
ax.text(x_line[-1]+0.25, y_line4[-1], 'Line four', color='g', fontsize=14)
ax.set_title('Labelling with Text')
#ax.set_xlabel('x axis')
#ax.set_ylabel('y axis')

ax = matplotlib.pyplot.subplot(3,2,5) ##rows, columns, plot
ax.semilogx(x_scatter, y_scatter, 'ro', ms=5, label='Scatter 1')
#ax.semilogx(x_scatter, z_scatter, 'gv', ms=5, label='Scatter 2')
legend = ax.legend()
legend.get_frame().set_facecolor('PowderBlue')
ax.set_title('Labelling with Legend, a pointless case')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax = matplotlib.pyplot.subplot(3,2,6) ##rows, columns, plot
ax.semilogy(x_scatter, y_scatter, 'ro', ms=5)
ax.semilogy(x_scatter, z_scatter, 'go', ms=5)
ax.set_title('Labelling with Text')
ax.set_xlabel('x axis')
#ax.set_ylabel('y axis')
ax.text(0.1, 5e-3, 'Scatter 1', color='r', fontsize=14)
ax.text(0.1, 2e-3, 'Scatter 2', color='g', fontsize=14)
#ax.legend()
```

```python
###Plot showing possible positions of legend
matplotlib.pyplot.figure(figsize=(8,5)) ##width, height

ax = matplotlib.pyplot.subplot(111)
ax.axis([0,1,0,1]) #xmin, xmax, ymin, ymax
ax.plot()
ax.set_title('Legend locations (loc=#)')

###Legend location by number
ax.text(0.9,0.9, '1', fontsize=16)
ax.text(0.1,0.9, '2', fontsize=16)
ax.text(0.1,0.1, '3', fontsize=16)
ax.text(0.9,0.1, '4', fontsize=16)
ax.text(0.1,0.5, '6', fontsize=16)
ax.text(0.85,0.5, '5 & 7', fontsize=16)
ax.text(0.5,0.1, '8', fontsize=16)
ax.text(0.5,0.9, '9', fontsize=16)
ax.text(0.5,0.5, '10', fontsize=16)

###Making the axes/spines thicker
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(1.5)

###Setting the fontsize of axes tick labels
ax.tick_params(axis='both', labelsize=14, which='major') ##axis = 'x' | 'y' | 'both'; which = 'major' | 'minor' | 'both


###Uncomment this to try out. loc=0 is "best"
#ax.plot(x_line, x_line, 'b-', lw=3, label='Line')
#ax.legend(loc=0)
```

```python
###Plot showing text labelling: relative to axes and to user coordinates
matplotlib.pyplot.figure(figsize=(10,10)) ##width, height

ax1 = matplotlib.pyplot.subplot2grid((2, 2), (0, 0))
ax2 = matplotlib.pyplot.subplot2grid((2, 2), (0, 1), rowspan=2)
ax3 = matplotlib.pyplot.subplot2grid((2, 2), (1, 0)) #, colspan=2)

ax1.axis([0,10,3.0,6.5])
#ax1.axis([0,10,1.0,10.0])
ax1.plot(x_line, y_line2, 'b-', lw=3)
ax1.set_xticks([])
ax1.text(1.0, 6.0, 'Blue line', color='b', fontsize=13, horizontalalignment='left', style='italic')
ax1.text(0.9, 0.1, 'blue curve', color='b', fontsize=14, horizontalalignment='right', transform=ax1.transAxes)

ax2.axis([0,10,-1,7])
#ax2.axis([0,10,-2,10])
ax2.plot(x_line, y_line3, 'g-', lw=3)
ax2.plot(x_line, y_line2, 'b-', lw=3)
ax2.text(1.0, 6.0, 'Two lines', color='k', fontsize=13, weight='medium', style='oblique')
ax2.text(0.9, 0.5, 'two curves', color='k', fontsize=14, horizontalalignment='right', transform=ax2.transAxes)

ax3.axis([0,10,-1,1])
#ax3.axis([0,10,-2,2])
ax3.plot(x_line, y_line3, 'g-', lw=3)
ax3.text(5.0, -0.75, 'Green line', color='g', fontsize=13, weight='bold')
ax3.text(0.9, 0.9, 'cos(x)', color='g', fontsize=14, horizontalalignment='center', transform=ax3.transAxes)
```
